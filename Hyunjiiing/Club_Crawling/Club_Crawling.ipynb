```python
from bs4 import BeautifulSoup
import requests

res = {'title': [], 'date': [], 'url': []}

list=[]

html = requests.get(f"https://www.chungbuk.ac.kr/site/www/sub.do?key=817").text
bsObject = BeautifulSoup(html, "html.parser")

for link in bsObject.find_all('tr'):
    list.append(link)
    
for index, link in enumerate(list):
    print(index, link)
```

```python
list = list[2:48]
print(list)
```

```python
col = ['순서', '분과', '동아리명', '대표학생', '지도교수 성명', '지도교수 학과', '회원수', '최초등록일', '목표']
res = { '순서': [], '분과': [], '동아리명': [], '대표학생': [], '지도교수 성명': [], '지도교수 학과': [], '회원수': [], '최초등록일': [], '목표': [] }
for link in list:
    temp = link.find_all('td')
    index = 0
    for line in temp:
        if index == 1 and len(temp) != len(col):
            index += 1
        
        res[col[index]].append(line.text)
        index += 1
```

```python
res.pop('분과')
res.pop('최초등록일')
```
